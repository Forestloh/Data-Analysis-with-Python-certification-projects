```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# reading csv file
df = pd.read_csv('epa-sea-level.csv')
```

```python
# Looking at examples of data fields
df.head()
```

<font color='#9999ff'> **Task: Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.**</font>

```python
# Creating a figure and axis with matplotlib
fig, ax = plt.subplots(figsize=(9, 4))

ax.scatter(x=df['Year'], y=df['CSIRO Adjusted Sea Level'], label = 'data')

# set title, x and y labels
ax.set(title='Rise in Sea Level', xlabel='Year', ylabel='Sea Level (inches)')
```

<font color='#9999ff'> **Task: Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit.**</font>

Calculate regression

```python
reg = linregress(x=df['Year'], y=df['CSIRO Adjusted Sea Level'], alternative='two-sided')
```

<font color='#9999ff'> **Task: Plot the line of best fit over the top of the scatter plot.**

**Make the line go through the year 2050 to predict the sea level rise in 2050.**</font>

Create list of years from earliest up till 2050

```python
# Find earliest year value in df
yr_min = df['Year'].min()

# create to store year values up till 2050
years = pd.Series([yr for yr in range(yr_min, 2051)])
```

Plot a regression line for year 1880 till 2050

```python
ax.plot(years, reg.intercept + reg.slope*years,color='red', label = 'reg line: yr 1880 to 2050')
ax.legend()
```

<font color='#9999ff'> **Task: Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset.**

**Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.**</font>

Get data from year 2000

```python
df2 = df[df['Year']>= 2000]
```

Calculate regression

```python
reg2 = linregress(x=df2['Year'], y=df2['CSIRO Adjusted Sea Level'], alternative='two-sided')
```

```python
# create to store year values up till 2050
years2 = pd.Series([yr for yr in range(2000, 2051)])
```

Plot a regression line for year 2000 till 2050

```python
ax.plot(years2, reg2.intercept + reg2.slope*years2,color='purple', label = 'reg line: yr 2000 to 2050')
ax.legend()
```
